## Creation of few new fields based on EDA

```python
# video engagment score
df1=train.groupby(['video_id'])['engagement_score'].mean()
video_eng=[]
for i in train['video_id']:
    video_eng.append(df1[i])
train['video_eng']=video_eng

video_eng=[]
for i in test['video_id']:
    video_eng.append(df1[i])
test['video_eng']=video_eng
```

```python
#user id & category id
df1=train.groupby(['user_id','category_id'])['engagement_score'].mean()
cu_eng=[]
for index, row in train.iterrows():
    cu_eng.append(df1[row['user_id']][row['category_id']])
train["cu_eng"]=cu_eng

cu_eng=[]
for index, row in test.iterrows():
    cu_eng.append(df1[row['user_id']][row['category_id']])
test["cu_eng"]=cu_eng
```

```python
#user engagment score
df1=train.groupby(['user_id'])['engagement_score'].mean()
u_eng=[]
for index, row in train.iterrows():
    u_eng.append(df1[row['user_id']])
train["u_eng"]=u_eng

u_eng=[]
for index, row in test.iterrows():
    u_eng.append(df1[row['user_id']])
test["u_eng"]=u_eng
```

```python
#category  engagment score
df1=train.groupby(['category_id'])['engagement_score'].mean()
c_eng=[]
for index, row in train.iterrows():
    c_eng.append(df1[row['category_id']])
train["c_eng"]=c_eng

c_eng=[]
for index, row in test.iterrows():
    c_eng.append(df1[row['category_id']])
test["c_eng"]=c_eng
```

```python
#Binning of age

train['Age Groups'] = pd.qcut(
   train['age'], 
   [0, 0.2, 0.4,0.6, 0.8, 1], 
   labels=['0-20%', '21-40%', '41-60%', '61-80','81-100%']
)

test['Age Groups'] = pd.qcut(
   test['age'], 
   [0, 0.2, 0.4,0.6, 0.8, 1], 
   labels=['0-20%', '21-40%', '41-60%', '61-80','81-100%']
)
```

### One hot encoding

```python
from sklearn.preprocessing import OneHotEncoder
onehotencoder = OneHotEncoder()
def onehotencoder1(data,li):
    for j in li:
        X = onehotencoder.fit_transform(data[j].values.reshape(-1,1)).toarray()
        print(X.shape)
        dfOneHot = pd.DataFrame(X, columns = [j+"_"+str(int(i)) for i in range(X.shape[1])]) 
        df = pd.concat([data, dfOneHot], axis=1)
        df= df.drop([j], axis=1) 
        data=df
        print(df.head())
    return df
```

```python
li=["gender","profession","Age Groups"]
train=onehotencoder1(train,li)
test=onehotencoder1(test,li)
```
